```python
from google import genai
```

```python
api_key = input("Enter your API KEY:")
client = genai.Client(api_key=api_key)
```

```python
while True:

    question = input("Enter your mental health problem (or type 'exit' to quit): ")

    if question.lower() == "exit":
        print("Good Bye!")
        break

    prompt = f"""
You are a Mental Health Support Chatbot.

Rules:
- Answer ONLY mental health questions.
- If the question is not about mental health, reply:
  Sorry! I can answer only mental health related questions.
- Reply in the SAME language as the user.
- If the user writes in Roman Urdu, reply ONLY in Roman Urdu (English letters).
- NEVER reply in Hindi script or Urdu script.
- Keep your answer short, kind and supportive.
- End every answer with:
  This information is for educational purposes only. Please consult a mental health professional if needed.

User Question:
{question}
"""

    answer = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    print("\nBot:")
    print(answer.text)
    print("-" * 40)
```
