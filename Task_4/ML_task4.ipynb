```python
from google import genai
```

```python
api_key = input("Enter your API key: ")
client = genai.Client(api_key= api_key)
```

```python
while True:

    question = input("Ask your question (type exit to quit): ")

    if question.lower() == "exit":
        print("Good Bye")
        break

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=f"""
You are a Health Chatbot.

Rules:

1. If the question is not related to health, medicine, diseases, symptoms, doctors, hospitals, fitness or nutrition, reply only:
Sorry! I can answer only health-related questions.

2. If the question is related to health:
- Reply in the exact language used by the user.
- If the user writes in Roman Urdu, reply ONLY in Roman Urdu using English letters.
- Do NOT use Hindi script.
- Do NOT use Urdu script.
- Keep the answer short and simple.
- At the end write:
This information is for educational purposes only. Please consult a doctor.

Question:
{question}
"""
    )

    print("\nBot:\n")
    print(response.text)
    print("------------------------------------------")
```
